```python
!pip install git+https://github.com/neelnanda-io/Easy-Transformer.git

import torch
import easy_transformer
from easy_transformer import EasyTransformer
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using {device} device")
torch.set_grad_enabled(False)

model = EasyTransformer.from_pretrained('gpt2').to(device)
print("Loaded model")

prompt = "Don't mention the"
prompt_tokens = model.to_tokens(prompt, prepend_bos=False)
print(prompt_tokens)
print(prompt_tokens.shape)

stuff = model(prompt_tokens)
print(stuff.shape)
logits = stuff[0,-1,:]
probs = torch.nn.functional.softmax(logits, dim=-1)
print([model.tokenizer.decode(t) for t in prompt_tokens[0]])
sorted_probs, indices = probs.sort(descending=True)
for i in range(10):
    token = model.tokenizer.decode(indices[i])
    print(f'{sorted_probs[i]} -- {token}')
```
